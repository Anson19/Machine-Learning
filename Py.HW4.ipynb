```python
def CalProb(train_data,train_label): 
    num,dimsnum = train_data.shape
    labelnum = len(np.unique(train_label)) 
    pj = np.zeros((labelnum))
    pji1 = np.zeros((labelnum,dimsnum))
    for i in range(num):
        label = train_label[i][0]
        pj[label] = pj[label]+1  
        for j in range(dimsnum):
            pji1[label][j] = pji1[label][j] + train_data[i][j] 
    
    pji1 = (pji1.T + 1) / (pj + 2) 
    pj = (pj + 1) / (num + labelnum) 
    return  pj, pji1 

```

```python
def TestProb_xi_j(xi,pji): 
    return xi * np.log(pji) + (1-xi) * np.log(1-pji)
```

```python
p_j,p_ji1,=CalProb(x_train,y_train)
```

```python
def classfier(test_data,test_label,pji1,pj):  
    row,column = test_data.shape
    labelnum = len(np.unique(test_label))
    acc = 0
    pred=[]
    for i in range(row):
        testdata = test_data[i,:]
        p_yj_xi = np.log(pj) 
        for j in range(labelnum): 
            for k in range(column):
                xk = testdata[k] 
                p_yj_xi[j] = p_yj_xi[j] + TestProb_xi_j(xk,pji1[j][k])
   
        p_y_xi = np.argmax(p_yj_xi)
        acc = acc + (p_y_xi == test_label[i])
        #print('real is: ',test_label[i],'  predict is: ',p_y_xi)
        pred.append(p_y_xi)
    print('Test accuracy is: ', acc/row)
    return(pred)
```

```python
data=load_breast_cancer()
len(data)
cancer_data = data['data']
cancer_target = data['target']
cancer_names = data['feature_names']
cancer_desc = data['DESCR']
X_train, X_test, y_train, y_test = train_test_split(cancer_data,cancer_target, test_size=0.4, random_state=30)
X_train, X_valid, y_train, y_valid = train_test_split(cancer_data,cancer_target,test_size=1/3,random_state=30)

```

```python
colors = ['navy', 'turquoise', 'darkorange']

def make_ellipses(gmm, ax):
    for n, color in enumerate(colors):
        if gmm.covariance_type == 'full':
            covariances = gmm.covariances_[n][:1, :1]
        elif gmm.covariance_type == 'tied':
            covariances = gmm.covariances_[0:1, 0:1]
        elif gmm.covariance_type == 'diag':
            covariances = np.diag(gmm.covariances_[n][0:1,0])
        elif gmm.covariance_type == 'spherical':
            covariances = np.eye(gmm.means_.shape[1]) * gmm.covariances_[n]
        v, w = np.linalg.eigh(covariances)
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[0], u[0])
        angle = 180 * angle / np.pi  # convert to degrees
        v = 2. * np.sqrt(2.) * np.sqrt(v)
        ell = mpl.patches.Ellipse(gmm.means_[n, :0], v[0], v[0],
                                  180 + angle, color=color)
        ell.set_alpha(0.5)
        ax.add_artist(ell)
```

```python
clf = GaussianMixture(n_components=2, covariance_type='tied')  

clf.means_init = np.array([X_train[y_train == i].mean(axis=0)
                                for i in range(2)])
# fit
clf.fit(X_train, y_train)
# predict
pred_train = clf.predict(X_train)
pred = clf.predict(X_test)
#evaluate
print ('Train accuracy = ' + str(accuracy_score(y_train, pred_train)))
print ('Test accuracy = ' + str(accuracy_score(y_test, pred)))
```
